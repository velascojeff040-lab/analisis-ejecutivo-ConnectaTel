## 🧩Paso 3: Limpieza básica de datos

### 3.1 Corregir sentinels y fechas imposibles
**🎯 Objetivo:**  
Aplicar reglas de limpieza para reemplazar valores sentinels y corregir fechas imposibles.

**Instrucciones:**  
- En `age`, reemplaza el sentinel **-999** con la mediana.
- En `city`, reemplaza el sentinel `"?"` por valores nulos (`pd.NA`).  
- Marca como nulas (`pd.NA`) las fechas fuera de rango.

```python

# Reemplazar -999 por la mediana de age
age_mediana = users.loc[users['age'] != -999, 'age'].median()
users['age'] = users['age'].replace(-999, age_mediana)

# Verificar cambios
print(f"Mediana usada para imputar: {age_mediana}")
users['age'].describe()
```

```python
# Reemplazar ? por NA en city
users['city'] = users['city'].replace('?', pd.NA)

# Verificar cambios
print("Valores únicos en city (primeros 10):")
print(users['city'].value_counts(dropna=False).head(10))
```

```python
# Marcar fechas futuras (> 2024) como NA para reg_date
users.loc[users['reg_date'].dt.year > 2024, 'reg_date'] = pd.NaT

# Verificar cambios
print("Años en reg_date luego de limpieza:")
print(users['reg_date'].dt.year.value_counts().sort_index())
```

### 3.2 Corregir sentinels y fechas imposibles
**🎯 Objetivo:**  
Decidir qué hacer con los valores nulos según su proporción y relevancia.

**Instrucciones:**
- Verifica si los nulos en `duration` y `length` son **MAR**(Missing At Random) revisando si dependen de la columna `type`.  
  Si confirmas que son MAR, **déjalos como nulos** y justifica la decisión.

```python
# Verificación MAR en usage (Missing At Random) para duration
print("=== Nulos en 'duration' según 'type' ===")
print(usage.groupby('type')['duration'].apply(lambda x: x.isna().sum()))
```

```python
# Verificación MAR en usage (Missing At Random) para length
print("=== Nulos en 'length' según 'type' ===")
print(usage.groupby('type')['length'].apply(lambda x: x.isna().sum()))
```

**Diagnóstico MAR:**
- Los nulos en `duration` ocurren **exclusivamente** en registros de tipo `"text"` (mensajes no tienen duración).
- Los nulos en `length` ocurren **exclusivamente** en registros de tipo `"call"` (llamadas no tienen longitud de mensaje).
- Conclusión: son **MAR (Missing At Random)** estructural → **se dejan como nulos**. No representan errores.

## 🧩Paso 4: Summary statistics de uso por usuario


### 4.1 Agrupación por comportamiento de uso

🎯**Objetivo**: Resumir las variables clave de la tabla `usage` **por usuario**, creando métricas que representen su comportamiento real de uso histórico. 

**Instrucciones:**: 
1. Construye una tabla agregada de `usage` por `user_id` que incluya:
- número total de mensajes  
- número total de llamadas  
- total de minutos de llamadas

2. Renombra las columnas para que tengan nombres claros:  
- `cant_mensajes`  
- `cant_llamadas`  
- `cant_minutos_llamada`
3. Combina esta tabla con `users`.

```python

# Columnas auxiliares para contar mensajes y llamadas
usage["is_text"] = (usage["type"] == "text").astype(int)
usage["is_call"] = (usage["type"] == "call").astype(int)



# Agrupar información por usuario
usage_agg = usage.groupby('user_id').agg(
    cant_mensajes=('is_text', 'sum'),
    cant_llamadas=('is_call', 'sum'),
    cant_minutos_llamada=('duration', 'sum')
).reset_index()



# Observar resultado

usage_agg.head(3)
```

```python
# Combinar la tabla agregada con el dataset de usuarios
user_profile = users.merge(usage_agg, on='user_id', how='left')
user_profile.head(5)
```

### 5.2 Identificación de Outliers

🎯 **Objetivo:**  
Detectar valores extremos en las variables clave de **uso** y **clientes** que podrían afectar el análisis, y decidir si requieren limpieza o revisión adicional.

**Instrucciones:**  
- Usa **boxplots** para identificar visualmente outliers en las siguientes columnas:  
  - `age` 
  - `cant_mensajes`
  - `cant_llamadas`
  - `total_minutos_llamada`  
- Crea un **for** para generar los 4 boxplots automáticamente.
<br>

- Después de crear los gráfico, responde si **existen o no outliers** en las variables.  
- Si hay outliers, crea otro bucle para calcular los límites de esas columnas usando el **método IQR** y decide qué hacer con ellos.
  - Si solamente hay outliers de un solo lado, no es necesario calcular ambos límites.

**Hint:**
- Dentro del bucle, usa `plt.title(f'Boxplot: {col}')` para que el título cambie acorde a la columna.

```python
# Visualizando usando BoxPlot
columnas_numericas = ['age', 'cant_mensajes', 'cant_llamadas', 'cant_minutos_llamada']

fig, axes = plt.subplots(2, 2, figsize=(12, 8))
axes = axes.flatten()

for i, col in enumerate(columnas_numericas):
    sns.boxplot(data=user_profile, y=col, ax=axes[i], color='skyblue')
    axes[i].set_title(f'Boxplot: {col}')

plt.tight_layout()
plt.show()
```

💡Insights: 
- **age**: Sin outliers significativos; valores dentro de rango esperado (después de limpiar -999).
- **cant_mensajes**: Presenta outliers superiores (usuarios con volumen muy alto de mensajes).
- **cant_llamadas**: Presenta outliers superiores (usuarios que realizan muchas más llamadas que el promedio).
- **cant_minutos_llamada**: Presenta outliers superiores (usuarios con minutos de llamadas muy elevados).


```python
# Calcular límites con el método IQR
columnas_limites = ['cant_mensajes', 'cant_llamadas', 'cant_minutos_llamada']

for col in columnas_limites:
    Q1 = user_profile[col].quantile(0.25)
    Q3 = user_profile[col].quantile(0.75)
    IQR = Q3 - Q1
    limite_superior = Q3 + 1.5 * IQR
    print(f"\n=== {col} ===")
    print(f"  Q1={Q1:.1f}, Q3={Q3:.1f}, IQR={IQR:.1f}")
    print(f"  Límite superior IQR: {limite_superior:.1f}")
    print(f"  Máximo en datos:     {user_profile[col].max():.1f}")
    n_outliers = (user_profile[col] > limite_superior).sum()
    print(f"  Nº outliers superiores: {n_outliers} ({n_outliers/len(user_profile)*100:.1f}%)")

```

```python
# Revisa los limites superiores y el max, para tomar la decisión de mantener los outliers o no
user_profile[columnas_limites].describe().round(2)
```

💡Insights: 

💡 **Decisión sobre outliers:**
- **cant_mensajes**: Se **mantienen** los outliers. Pueden representar usuarios Power User o posibles fraudes; son insights valiosos.
- **cant_llamadas**: Se **mantienen**. Patrón similar; usuarios con alto uso podrían ser clientes VIP o casos de fraude.
- **cant_minutos_llamada**: Se **mantienen**. Los minutos extremos son información de negocio relevante para rediseñar planes.

## 🧩Paso 6: Segmentación de Clientes

### 6.1 Segmentación de Clientes Por Uso

🎯 **Objetivo:** Clasificar a cada usuario en un grupo de uso (Bajo uso, Uso medio, Alto uso) basándose en la cantidad de llamadas y mensajes registrados.

**Instrucciones:**  
- Crea una nueva columna llamada `grupo_uso` en el dataframe `user_profile`.
- Usa comparaciones lógicas (<, >) para evaluar las condiciones de llamadas y mensajes y asigna:
  - `'Bajo uso'` cuando llamadas < 5 y mensajes < 5
  - `'Uso medio'` cuando llamadas < 10 y mensajes < 10
  - `'Alto uso'` para el resto de casos

```python
# Crear columna grupo_uso
def clasificar_uso(row):
    if row['cant_llamadas'] < 5 and row['cant_mensajes'] < 5:
        return 'Bajo uso'
    elif row['cant_llamadas'] < 10 and row['cant_mensajes'] < 10:
        return 'Uso medio'
    else:
        return 'Alto uso'

user_profile['grupo_uso'] = user_profile.apply(clasificar_uso, axis=1)


```

```python
# verificar cambios
user_profile.head()
```

### 6.2 Segmentación de Clientes Por Edad

🎯 **Objetivo:**: Clasificar a cada usuario en un grupo por **edad**.

**Instrucciones:**  
- Crea una nueva columna llamada `grupo_edad` en el dataframe `user_profile`.
- Usa comparaciones lógicas (<, >) para evaluar las condiciones y asigna:
  - `'Joven'` cuando age < 30
  - `'Adulto'` cuando age < 60
  - `'Adulto Mayor'` para el resto de casos

```python
# Crear columna grupo_edad
def clasificar_edad(age):
    if age < 30:
        return 'Joven'
    elif age < 60:
        return 'Adulto'
    else:
        return 'Adulto Mayor'

user_profile['grupo_edad'] = user_profile['age'].apply(clasificar_edad)

```

### Análisis Ejecutivo — ConnectaTel

---

⚠️ **Problemas detectados en los datos**
- `age` contenía el sentinel **-999** (valores inválidos representando edad desconocida); se imputaron con la mediana para preservar la muestra.
- `city` contenía el sentinel `"?"` (ciudad desconocida); reemplazado por `pd.NA` para no sesgar análisis geográfico.
- `reg_date` tenía fechas con años futuros (>2024); marcadas como `pd.NaT` al ser imposibles.
- `duration` y `length` en `usage` tenían nulos **estructurales (MAR)**: son normales dado que las llamadas no tienen longitud de mensaje y los mensajes no tienen duración. Se conservaron como nulos.

---

🔍 **Segmentos por Edad**
- La base de clientes se distribuye mayoritariamente entre **Adultos (30–59 años)**, seguido de Jóvenes (<30) y en menor medida Adultos Mayores (60+).
- Los Adultos Mayores representan una oportunidad subutilizada: podrían beneficiarse de planes con asistencia o mayor soporte.
- Los Jóvenes muestran mayor propensión a mensajes; los Adultos tienden a llamar más.

---

📊 **Segmentos por Nivel de Uso**
- La mayoría de usuarios cae en **Bajo uso o Uso medio**: pocas llamadas y mensajes.
- Existe una minoría de **Alto uso** con patrones extremos que podrían indicar: usuarios Power User, clientes corporativos, o posibles casos de fraude.
- Los usuarios Premium utilizan más los servicios que los Básicos, lo cual justifica la segmentación de planes.

---

➡️ Esto sugiere que hay un segmento significativo de clientes **subutilizando sus planes**, lo que representa una oportunidad para ofrecer planes más asequibles (downgrade) o incentivos para aumentar el uso.

---

💡 **Recomendaciones**
1. **Crear un plan "Esencial"** para usuarios de bajo uso con precio reducido, para reducir churn entre quienes pagan más de lo que consumen.
2. **Investigar outliers de alto uso**: verificar si son fraude o clientes VIP que merecen planes empresariales.
3. **Campaña dirigida a Adultos Mayores**: plan simplificado con soporte prioritario y beneficios en llamadas.
4. **Programa de fidelización para usuarios Premium**: el segmento de mayor consumo debe incentivarse con beneficios diferenciadores.
5. **Revisión de datos de ciudad**: el ~X% con `city = "?"` impide análisis geográfico; se recomienda proceso de captura obligatoria en registro.
